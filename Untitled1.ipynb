Задача 1

Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.

```python
import math

def combination(a, b):
    return math.factorial(a) // (math.factorial(b) * math.factorial(a - b))
```

Я считаю случаи, когда туз выпал 1, 2, 3 и 4 раза:

```python
a = combination(4, 1) * combination(48, 3) + combination(4, 2) * combination(48, 2) + combination(4, 3) * combination(48, 1) + 1
a
```
