Контрольная работа состоит из пяти вопросов. На каждый вопрос приведено четыре варианта ответа, один из которых правильный. Случайная величина X задаёт число правильных ответов при простом угадывании. Найдите математическое ожидание данной случайной величины. Найдите дисперсию случайной величины X.

Подсказка: постройте закон распределения случайной величины X.

```python
import math

def combination(a, b):
    return math.factorial(a) // (math.factorial(b) * math.factorial(a - b))
```

Студент может ответить X = 0, 1, 2, 3, 4 и 5 раз правильно (случайная величина)

```python
p = 1/4 # вероятность ответить правильно
q = 3/4 # вероятность ответить неправильно
```

Вероятность при X = 0 равна:

```python
X_0 = (3/4) ** 5
```

Вероятность при X = 1 по формуле Бернулли равна:

```python
X_1 = combination(5, 1) * p * q ** 4
```

И т.д.

```python
X_2 = combination(5, 2) * p ** 2 * q ** 3
X_3 = combination(5, 3) * p ** 3 * q ** 2
X_4 = combination(5, 4) * p ** 4 * q ** 1
X_5 = combination(5, 5) * p ** 5 * q ** 0
```

Распределение равно: 

```python
print(X_0, X_1, X_2, X_3, X_4, X_5)
```

```python
M_X = 5 * p # Мат ожидание
D_X = 5 * p * q # Дисперсия
```

```python
print(M_X, D_X)
```

## Задача 3
Пользователь получает в среднем 10 писем со спамом на свой почтовый ящик за сутки. Найти число N, такое, что с вероятностью 0.95 пользователь получит не более N писем со спамом за текущий день.

```python
import numpy as np
def poisson_proba(k: int, lambda_: float) -> float:
    """Формула Пуассона.
    """
    
    return (lambda_ ** k) * (np.exp(-lambda_)) / np.math.factorial(k)
```

```python
lambda_ = 10
```

Сумма вероятностей должна быть равна меньше 0,95, тогда:

```python
i = 0
my_sum = 0
while (my_sum < 0.95):
    my_sum += poisson_proba(k=i, lambda_=lambda_)
    i += 1
i - 1
```
